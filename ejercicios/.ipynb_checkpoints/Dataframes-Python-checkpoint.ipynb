```python
letras = ["A", "B", "C"]
data1 = pd.DataFrame({"Valor-1": np.random.randint(1, 10, 10), "Valor-2": np.random.randint(1, 10, 10), "Key": np.random.choice(letras, 10)})
data1
```

## Agregados 

```python
data1.groupby("Key")["Valor-1"].aggregate(suma=sum, minimo=np.min, media=np.mean, maximo=np.max, 
                                          mediana=np.median, std=np.std, varianza=np.var)
```

```python
data1.groupby("Key").agg([sum, np.min, np.mean, np.max])
```

## Carga DataFrame `Titanic`

```python
titanic = pd.read_csv("../data/titanic.csv")
```

```python
titanic.shape
```

```python
titanic.head()
```

## Reemplazar valores `NaN` por la media

```python
edad_media = np.round(np.mean(titanic.Age),0)
titanic["Age"].fillna(edad_media, inplace=True)
```

```python
np.random.seed(12512)
muestra = np.random.randint(0, 1313, 10)
muestra = np.sort(muestra)
titanic.iloc[muestra]
```

```python
titanic.Survived.value_counts()
```

```python
titanic.shape
```

```python
titanic[titanic.Age > 30].sort_values("Age").tail(5)
```

## Eliminar Duplicados

Elimina elementos duplicados seg√∫n el parametro `Name`

```python
titanic_dupl = titanic.drop_duplicates(subset="Name")
titanic_dupl.shape[0]
```

## GroupBy

Mostrar los individuos que sobrevivieron y no sobrevivieron en el Titanic

```python
titanic.groupby("Survived")["Name"].count()
```

Calcular la media de la edad de los individuos agrupadas por el Sexo y si sobrevivieron o no

```python
np.round(titanic.groupby(["Sex", "Survived"])["Age"].mean(),2)
```

Mostrar los estad√≠sticos de resumen de la Edad de los pasajeros del Titanic agrupados por el g√©nero

```python
titanic.groupby("Sex").agg(minimo=("Age", np.min), 
                           media=("Age", np.mean), 
                           maximo=("Age", np.max),
                           mediana=("Age", np.median),
                           std=("Age", np.std),
                           varianza=("Age", np.var))
```
