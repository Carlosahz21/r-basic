# Gráficas y sus parámetros

Para crear una grafica generica se usa la funcion `plt.plot()`, para agregar parámetros a una gráfica de tipo `plt.plot()` para mostrar la gráfica siempre se ingresa `plt.show()`, se pueden usar los parámetros definidos en la documentacion del modulo `matplotlib.lines.Line2D`, hay parámetros interesantes como la forma de los puntos, para cambiar la forma de los puntos se modifica el parámetro `marker` y su documentación se encuentra en el módulo `matplotlib.markers`

```python
np.random.seed(125)
x = np.arange(1,21)
y = np.random.randint(1,11, len(x)) #genera varios numeros enteros aleatorios entre 1 y 10
plt.figure(figsize=(8, 6), dpi=80) #Tamaño de la gráfica
plt.plot(x, y,               #Ejes de la gráfica 
         marker="H",        #tipo de figura en cada punto, se puede usar LATEX
         ms=12,            #ms = markersize, es el tamaño de la figura de los puntos
         color="orange",   #color de la linea
         mew=3,            #mew = markeredgewidth,es el ancho de la linea del borde de la figura en los puntos
         mec="blue",       #mex = markeredgecolor, es el color del borde la figura en los puntos
         mfc="red",        #mfc = markerfacecolor, es el color del centro de la figura en los puntos
         linewidth=2,      #ancho de la linea
         linestyle=":")    #estilo de linea, ":" en este caso hace una linea de puntos

plt.show()
```

# Parametros Globales de una Gráfica

Para modificar lo  parametros globales de una gráfica por defecto se usa la función `plt.setp()` que significa **set properties**

```python
np.random.seed(216)
formas = ["o", "v", "^", "D", "H", "p", "<", ">", "3", "8", "P", "*", "."]
alphas = np.round(np.random.rand(10),1)
x = {}
y = {}
plt.figure(figsize=(8, 6), dpi=80)
for i in range(20):  
    x[i] = 9*np.random.rand(10)     #genera i listas de numeros aleatorios en el dicionario x
    y[i] = 11*np.random.rand(10)    #genera i listas de numeros aleatorios en el dicionario y
    figura = plt.plot(x[i], y[i], marker=np.random.choice(formas), ms=8, ls="", alpha=np.random.choice(alphas)) 
    plt.setp(figura, marker=np.random.choice(formas))

plt.title("Scatter Plot")
plt.show()
```

```python
plt.setp(figura)
```

# Sublots 

función: 
$$f(x)= e^{-x} \cdot \cos(2\pi x)$$

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)

x1 = np.arange(0, 5, 0.1)
x2 = np.arange(0, 5, 0.02)

plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(211) #los 2 primeros digitos son el aspect ratio del grid 2:1
plt.plot(x1, f(x1), "ro", x2, f(x2), "k")

plt.subplot(212)
plt.plot(x2, f(x2), 'g:')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0.0, 2.0, 201)
s = np.sin(2 * np.pi * t)

#Color del fondo del todo el marco
fig, axs = plt.subplots(1, 2, facecolor="#B2C99A", figsize=(9, 5), dpi=820)
#Color del Grid
axs[0].set_facecolor('#eafff5')
axs[0].set_title('Voltage $vs$ time chart', color='#56686F')
axs[0].set_xlabel('time ($s$)', color="#482BE1")
axs[0].set_ylabel('voltage ($mV$)', color='#1F170E')
axs[0].plot(t, s, 'red')
axs[0].plot(t, .7*s, color='C4', linestyle='-.')
axs[0].tick_params(labelcolor='grey')
axs[0].grid(True)

axs[1].plot([1,2,3,4],[1,0,1,0])

plt.tight_layout()
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

dt = 0.01
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t)) # white noise 1
nse2 = np.random.randn(len(t)) # white noise 2

# Two signals with a coherent part at 10Hz and a random part
s1 = np.sin(2*np.pi*10*t) + nse1
s2 = np.sin(2*np.pi*10*t) + nse2

fig, axs = plt.subplots(2, 1,figsize=(8, 6), dpi=80)
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('time')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, 256, 1. / dt)
axs[1].set_ylabel('coherence')

fig.tight_layout()
plt.show()
```

# Textos en Gráficos

El módulo `scipy.stats` tiene instancias de las districuciones de probabilidad mas importantes, en este caso se va a utilizar la distribucion normal

```python
from scipy.stats import norm
np.random.seed(587)

#Media y desviacion Estandar
mu = 100 
sigma = 25

#Vector Aleatorio con distribucion normal
x = mu + sigma*np.random.randn(10000)
x = np.sort(x)

#Instanciación de la distribucion Normal
#con media = mu, y desviacion estandar = sigma 
rv = norm(mu, sigma)
#Así se calculan los primeros 4 momentos de probabilidad
mean, var, skew, kurt = rv.stats(moments='mvsk')
#mean = media, var = varianza,
#skew = coeficiente de simetria
#kurt = curtosis
```

```python
plt.figure(figsize=(8, 6), dpi=80)
#Historgrama
n, bins, patches = plt.hist(x, 15, density=1, facecolor="#5581AD",alpha=0.7) 
#Gráfica de la curva normal sobre el Histrograma
plt.plot(np.sort(x), rv.pdf(x), color = "#f35d52", lw = 3)
plt.title("Histograma CI")
plt.xlabel("Cociente Intelecual")
plt.ylabel("Probabilidad")
plt.text(137.5, 0.013, r"$\mu=100,\ \sigma=5$")
plt.grid(True)

plt.savefig("histograma.png")
plt.show()
```

# Agregar una leyenda a un Gráfico

```python
import numpy as np
import matplotlib.pyplot as plt

plt.subplots(figsize=(8,5), dpi=300)
# fig.set_facecolor("yellow")
# ax1.set_facecolor("grey")

# ejes
plt.xticks([-2*np.pi,-3*np.pi/2,-np.pi, -np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2,2*np.pi],
          ["$-2\pi$","$-\dfrac{3\pi}{2}$",'$-\pi$', '$-\dfrac{\pi}{2}$', '$0$', '$\dfrac{\pi}{2}$', '$\pi$', 
           "$\dfrac{3\pi}{2}$", "$2\pi$"])
plt.yticks(np.linspace(-1,1, 2, endpoint=True),
          ['-1',  '1'])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
# labels que van en la leyenda
ax.plot(x, y1, color='#D12642', label='Seno', lw=2, ls="-")
ax.plot(x, y2, color='#499775', label='Coseno', lw=2, ls="-")

# Crear leyenda
ax.legend(loc="best")

#punto en plano
ax.plot(-3*np.pi/2, 1, marker="H", ms=8, color = "orange")
ax.plot([-3*np.pi/2, -3*np.pi/2], [0, np.sin(-3*np.pi/2)], color="#F39B36", ls="-.")
ax.plot([0, -3*np.pi/2], [1, np.sin(-3*np.pi/2)], color="#F39B36", ls="-.")

# Anotacion sobre el punto
ax.set_ylim(-1.5,1.5)
ax.annotate(r"$\sin(-\dfrac{3\pi}{2})=1$",
           xy=(-3*np.pi/2,1),
           xytext=(-np.pi,1.25),
           fontsize=12,
           arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=0,angleB=90"))

# Guardar el grafico
plt.savefig('sin_cos_4.png')
```

# Cambio de Escala

```python
from matplotlib.ticker import NullFormatter

mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala Logarítmica")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x, y - y.mean())
plt.yscale("symlog", linthreshy=0.01)
plt.title("Escala Log Simétrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logística")
plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.grid(True)

plt.tight_layout()

plt.show()
```

# Graficos usando Bokeh

```python
from bokeh.plotting import figure, output_file, output_notebook, show

# prepare some data
x = np.arange(-5,5,0.05)
f = x**2
g = 5*x
h = (x**2)*np.sin(np.pi*x)

# output to static HTML file
#output_file("lineas.html")
output_notebook()
# create a new plot with a title and axis labels
p = figure(title="Funciones", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend and line thickness
p.line(x, f, legend_label="f(x)", line_width=3)
p.line(x, g, legend_label="g(x)",line_width=3, line_color="orange")
p.line(x, h, legend_label="h(x)", line_width=3, line_color="#9BE4A4")

# show the results
show(p)
```
