## Ejercicio 2

```python
plt.subplots(figsize=(8,6), dpi=150)
n = 1024 
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n) 

ax = plt.gca()
ax.set(xlim=(-1.5,1.5), ylim=(-1,1), xticks=[], yticks=[])

plt.scatter(x, y, s=80, c=np.arctan2(y,x), alpha=0.5)


plt.savefig("Ejercicio2 ")
```

## Ejercicio 3

```python
np.random.seed(1256)
plt.subplots(figsize=(8,6), dpi=150)

n = 12
x = np.arange(n)
y1 = (1 - x / float(n)) * np.random.uniform(0.5, 1.0, n)
y2 = (1 - x / float(n)) * np.random.uniform(0.5, 1.0, n) 

plt.bar(x, y1, facecolor='#9999ff', edgecolor='white')
plt.bar(x, -y2, facecolor='#ff9999', edgecolor='white') 

ax = plt.gca()
ax.set(ylim=(-1.25,1.25), xticks=[], yticks=[])

for xi, yi in zip(x, y1):     
    plt.text(xi, yi + 0.04, '%.2f' % yi, ha='center')

for xi, yi in zip(x, y2):     
    plt.text(xi, -yi - 0.11, '%.2f' % yi, ha='center')

plt.savefig("Ejericio3")  
```

## Ejercicio 4

```python
plt.subplots(figsize=(8,6), dpi=150)
n = 20
z = np.ones(n)
z[19] = 3

ax = plt.gca()
ax.set(ylim=(-1.25,1.25), xticks=[], yticks=[])

plt.pie(z, explode=z*.05, colors = ['%f' % (i/float(n)) for i in range(n)])
plt.axis('equal')

plt.savefig("Ejercicio4")
```
